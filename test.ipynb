```python
from bs4 import BeautifulSoup
import json
# Read the HTML file
import shutil
import os

# Get the current working directory
directory = os.getcwd()

all_reviews = []

id_ = 0

directory_prefix = "product4"

directory_name = f'{directory_prefix}/reviews_top'
# Iterate through all files in the directory
for filename in os.listdir(os.path.join(directory, directory_name)):
    # Check if the filename matches the pattern 'rXX.html'
    if filename.startswith('r') and filename.endswith('.html'):
        # Extract the digits XX from the filename
        # digits = filename[1:-5]
        # Perform any further processing you need
        # print("Found file:", filename)
        # print("Digits:", digits)
        print(filename)


        with open(directory_name+"/"+filename, 'r', encoding='utf-8') as file:
            html_content = file.read()

        # Parse the HTML
        soup = BeautifulSoup(html_content, 'html.parser')

        # print(soup)

        # Find all review containers
        review_containers = soup.find_all('div', class_='a-section review aok-relative')

        # print(review_containers)

        # Iterate over each review container
        for review in review_containers:

            
            review_data = {}
            # Extract reviewer name
            reviewer_name = review.find('span', class_='a-profile-name').text.strip()
            
            # Extract review rating
            review_rating = review.find('i', class_='a-icon-star').find('span', class_='a-icon-alt').text.strip()
            review_rate_no = int(float(review_rating.split(" ")[0]))
            
            # Extract review title
            review_title = review.find('a', class_='review-title').text.strip().split("\n")[-1]

            
            # Extract review date
            review_date = review.find('span', class_='review-date').text.strip()
            
            # Extract review body
            review_body = review.find('span', class_='review-text-content').text.strip()
            
            # Extract number of people found the review helpful
            # helpful_count = review.find('span', class_='cr-vote-text').text.strip()
            review_data = {
                "id": id_,
                "Reviewer": reviewer_name,
                "Rating": review_rating,
                "Title": review_title,
                "Date": review_date,
                "Body": review_body,
                "Number": review_rate_no
            }

            # Check if there are images
            if review.find('div', class_='review-image-tile-section'):
                # Extract image URLs/hrefs
                image_urls = [img['src'] for img in review.find_all('img', class_='review-image-tile')]

                for image_url in image_urls:
                    image_filename = os.path.basename(image_url)
                    print("!"*100)
                    path_1 = os.path.join(directory, ("/".join([directory_name] + image_url.split("/")[1:])))
                    path_2 = (os.path.join('r32_files', image_filename))
                    shutil.copy(path_1, path_2)
                    print(path_1)

                image_urls = [img.split("/")[-1] for img in image_urls]

                review_data['image_urls'] = image_urls
                

            all_reviews.append(review_data)
            id_ = id_ + 1
            print(review_data)
            
            # Print the extracted information
            # print("Reviewer:", reviewer_name)
            # print("Rating:", review_rating)
            # print("Title:", review_title)
            # print("Date:", review_date)
            # print("Body:", review_body)
            # # print("Helpful Count:", helpful_count)
            # print("\n")

review_json = json.dumps(all_reviews, indent=4)
file_path = f"{directory_prefix}/top_review_data.json"
with open(file_path, "w") as file:
    file.write(review_json)


```
